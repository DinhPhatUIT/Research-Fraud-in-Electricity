```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
input_file = 'E:/Phat/imbalance/data_electricity_process.parquet'
target_column = 'fraud'                   
output_train = 'df_train.parquet'
output_val = 'df_valid.parquet'
output_test = 'df_test.parquet'

df = pd.read_parquet('E:/Phat/imbalance/data_electricity_process.parquet')

df_train, df_temp = train_test_split(
    df,
    test_size=0.4,
    random_state=42,
    stratify=df[target_column]
)

df_val, df_test = train_test_split(
    df_temp,
    test_size=0.5,
    random_state=42,
    stratify=df_temp[target_column]
)

df_train.to_parquet(output_train, index=False)
df_val.to_parquet(output_val, index=False)
df_test.to_parquet(output_test, index=False)
```
